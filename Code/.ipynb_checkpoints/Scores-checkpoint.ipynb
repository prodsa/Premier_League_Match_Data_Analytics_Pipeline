Part 2 : Create a function that takes the DataFrame from Part 1 as an input and  

(1) Removes all rows where the fixture contains 'Man City'.

(2) Updates all cells where 'Brighton' occurs to 'Brighton & Hove Albion'.

(3) Returns a DataFrame that shows the standings for a given date.

Each team gets 3 points for a win, 1 for a draw and 0 for a loss.

```python
# Creating the above function: myfunc()
def myfunc(fix):
    print("\nInvoking function >>>")
    
    # (1) Remove all rows where the fixture contains 'Man City'
    a = fix[(fix.HomeTeam != 'Man City') & (fix.AwayTeam != 'Man City')]
    
    if 'Man City' not in a.values:
        print('\n1) Man City has been removed!')
    else:
        print('\n1) Man City still exists!')
    
    # (2) Update all cells where 'Brighton' occurs to 'Brighton & Hove Albion'
    for x in a.index:
        if a.loc[x, 'HomeTeam'] == "Brighton":
            a.loc[x, 'HomeTeam'] = "Brighton & Hove Albion"
    
        elif a.loc[x, 'AwayTeam'] == "Brighton":
            a.loc[x, 'AwayTeam'] = "Brighton & Hove Albion"
    
    if (('Brighton' not in a.values) and ('Brighton & Hove Albion' in a.values)):
        print("\n2) 'Brighton' has been updated to 'Brighton & Hove Albion'")
    else:
        print("\n2) 'Brighton' is not updated properly!")
    
    # (3) Returns a DataFrame that shows the standings for a given date.
    # Each team gets 3 points for a win, 1 for a draw and 0 for a loss.
    
    # Case 1 : When Home Team scores more than Away Team - Home Team gets 3 points and Away Team gets 0 points

    x = a[(a['HomeTeamGoals'] > a['AwayTeamScores'])]

    x['HomePoints'], x['AwayPoints'] = [3, 0]
    
    x1 = x[['Date', 'HomeTeam', 'HomeTeamGoals', 'AwayTeamScores', 'HomePoints']]
    x1.rename(columns={'HomeTeam' : 'Club', 'HomeTeamGoals' : 'GoalsScored', 'AwayTeamScores' : 'GoalsConceded', 'HomePoints' : 'Points'}, inplace=True)

    x2 = x[['Date', 'AwayTeam', 'AwayTeamScores', 'HomeTeamGoals', 'AwayPoints']]
    x2.rename(columns={'AwayTeam' : 'Club', 'AwayTeamScores' : 'GoalsScored', 'HomeTeamGoals' : 'GoalsConceded', 'AwayPoints' : 'Points'}, inplace=True)
    
    x3 = pd.concat([x1, x2])
    
    
    # Case 2 : When Away Team Scores more than Home Team - Away Team gets 3 points and Home Team gets 0 points
    
    y = a[(a['HomeTeamGoals'] < a['AwayTeamScores'])]
    
    y['HomePoints'], y['AwayPoints'] = [0, 3]
    
    y1 = y[['Date', 'AwayTeam', 'AwayTeamScores', 'HomeTeamGoals', 'AwayPoints']]
    y1.rename(columns={'AwayTeam' : 'Club', 'AwayTeamScores' : 'GoalsScored', 'HomeTeamGoals' : 'GoalsConceded', 'AwayPoints' : 'Points'}, inplace=True)

    y2 =6 y[['Date', 'HomeTeam', 'HomeTeamGoals', 'AwayTeamScores', 'HomePoints']]
    y2.rename(columns={'HomeTeam' : 'Club', 'HomeTeamGoals' : 'GoalsScored', 'AwayTeamScores' : 'GoalsConceded', 'HomePoints' : 'Points'}, inplace=True)
    
    y3 = pd.concat([y1, y2])
    
    
    # Case 3 : When Home Team draws with Away Team - Both Teams get 1 point each    
    
    z = a[(a['HomeTeamGoals'] == a['AwayTeamScores'])]

    z['Points'] = 1

    z1 = z[['Date', 'HomeTeam', 'HomeTeamGoals', 'AwayTeamScores', 'Points']]
    z1.rename(columns={'HomeTeam' : 'Club', 'HomeTeamGoals' : 'GoalsScored', 'AwayTeamScores' : 'GoalsConceded'}, inplace=True)

    z2 = z[['Date', 'AwayTeam', 'AwayTeamScores', 'HomeTeamGoals', 'Points']]
    z2.rename(columns={'AwayTeam' : 'Club', 'AwayTeamScores' : 'GoalsScored', 'HomeTeamGoals' : 'GoalsConceded'}, inplace=True)
    
    z3 = pd.concat([z1, z2])
    
    
    # Merging all three subset cases together:
    xyz = pd.concat([x3, y3, z3])
    
    # Converting 'Date' field from object to datetime format
    xyz['Date'] = pd.to_datetime(xyz['Date'], format='%d%m%Y', infer_datetime_format=True)
    
    # Sorting values by earliest date
    xyz.sort_values(by=['Date'], inplace=True)
    
    # Resetting the index
    xyz.reset_index(drop=True, inplace=True)
    
    
    # Taking the date as an input from the user to generate the standings report
    year = str(input('\nEnter a year (YYYY): '))
    
    # Premier league matches are held between Aug of current year to May of next year
    # Checking if month entered is not between June and July as this is the off season period
    
    while True:
        try:    
            month = str(input('\nEnter a month (MM): '))
            if int(month) < 8 and int(month) > 5:
                raise ValueError
            break
            
        except ValueError:
            print("\nMonth entered is incorrect. Enter month between Aug to May!")
    
    day = str(input('\nEnter a day (DD): '))
    
    if int(month) in range(1, 6):
        pre_year = str(int(year) - 1)
        start_date = pre_year + '-08-01'
        #print(start_date)
        
    else:
        start_date = year + '-08-01'
        #print(start_date)
    
    end_date = year + '-' + month + '-' + day

    print("\n3) Generating Standings Between " + start_date + " to " + end_date)
    
    
    # Creating a subset of all records between the start and end of season dates
    st = xyz[(xyz['Date'] >= start_date) & (xyz['Date'] <= end_date)]
    
    # Since 3 points are given if win, 1 point if draw and 0 point if loss, 
    # Setting 0 or 1 flag for the respective populated point value
    st['W'] = [1 if x == 3 else 0 for x in st['Points']]
    st['D'] = [1 if x == 1 else 0 for x in st['Points']]
    st['L'] = [1 if x == 0 else 0 for x in st['Points']]
    
    # Grouping by 'Club' based on sum 
    # To get total sum of goals scored, goals conceded, points, wins, draws and losses
    pos = st.groupby('Club').sum().reset_index()
    pos = pd.DataFrame(pos)

    # Grouping by 'Club' based on count or total number of records to get Number of Matches played 
    # Number of recurring Points records if any are totalled up and the column is renamed to 'MatchesPlayed'
    mp = st.groupby('Club').agg(MatchesPlayed=('Points', 'count')).reset_index()
    mp = pd.DataFrame(mp)
    
    # Merging the two dataframes using inner join based on common columns between the two
    standings = pd.merge(pos, mp)
    
    # GoalDifference (GD) can be calculated as the difference between GoalsScored & GoalsConceded
    standings['GoalDifference'] = standings['GoalsScored'] - standings['GoalsConceded']
    
    # Sorting the standings by the Team that has scored the highest points and also, resetting index
    standings = standings.sort_values(by=['Points'], ascending=False).reset_index(drop=True)
    
    # Creating a new column Position based on index to give each team a ranking position
    standings['Position'] = [x + 1 for x in standings.index]
    
    # Rearranging the order of the columns to fit in with the standings table 
    standings = standings[['Position', 'Club', 'MatchesPlayed', 'W', 'D', 'L', 'GoalsScored', 'GoalsConceded', 'GoalDifference', 'Points']]
    
    # Creating a mapping dictionary to map column names 
    # From standings dataframe attributes onto the snowflake table attributes 
    snowflake_mappings = {'Position' : 'POSITION', 
                          'Club' : 'CLUB',
                          'MatchesPlayed' : 'MATCHES_PLAYED',
                          'W' : 'WIN', 
                          'D' : 'DRAW',
                          'L' : 'LOSS',
                          'GoalsScored' : 'GOALS_FOR',
                          'GoalsConceded' : 'GOALS_AGAINST',
                          'GoalDifference' : 'GOAL_DIFFERENCE',
                          'Points' : 'POINTS'}

    # Renaming the standings dataframe attributes to match the snowflake table schema 
    standings.rename(columns = snowflake_mappings, inplace=True)
    
    print("\nRetrieving following Input DataFrame:")
    print(fix.to_string())
    print("\n")
    print(fix.info())
    
    print("\nOutput Standings Report is generated for: " + end_date + "\n")
    
    return standings
```

# Part 4

Using the Snowflake Connector, execute DELETE on the STANDINGS table to clear out any existing data

Then using myfunc() created in Part 2, write the Standings for the end of the season to this Snowflake table.

```python
# Importing the necessary packages for the snowflake connection

import snowflake.connector
import pandas
from snowflake.connector.pandas_tools import write_pandas
```

```python
context = 'C:/Users/prono/OneDrive/Desktop/Data_Engineering/Projects/Premier_League_Match_Data_Analytics_Pipeline/Source/ATT82825.env'

params = ['SNOWFLAKE_USERNAME', 
          'SNOWFLAKE_PASSWORD', 
          'SNOWFLAKE_ACCOUNT', 
          'SNOWFLAKE_WAREHOUSE', 
          'SNOWFLAKE_DATABASE', 
          'SNOWFLAKE_SCHEMA']

mydict = {}

# Opens the context file 'ATT82825.env' and 
# Searches every line for each attributes listed in 'params' list.

with open(context, encoding='utf-8') as f:
    for line in f:
        for p in params:
            if p in line:
                # Saves a substring value of 'line'
                # Index starts at the '=' sign and ends with the final character
                # leading and trailing spaces are also stripped off from val
                val = line[ (line.index('=') + 1) : len(line)].strip()
                mydict[p] = val
                
```

```python
mydict
```

```python
# Creating a connection to the Snowflake database table: TABLE DEVELOPER_SANDBOX.TECH_TEST.STANDINGS_13
# Specifying the database parameters here:

conn = snowflake.connector.connect(
    user=mydict['SNOWFLAKE_USERNAME'],
    password=mydict['SNOWFLAKE_PASSWORD'],
    account=mydict['SNOWFLAKE_ACCOUNT'],
    warehouse=mydict['SNOWFLAKE_WAREHOUSE'],
    database=mydict['SNOWFLAKE_DATABASE'],
    schema=mydict['SNOWFLAKE_SCHEMA']
)

cur = conn.cursor()
```

```python
try:
    # Checking snowflake connector version
    cur.execute("SELECT current_version()")
    one_row = cur.fetchone()
    print("\nSnowflake Connector version is:", one_row[0])
    
    # Retrives and displays all data from the table
    results = cur.execute("SELECT * from STANDINGS_13").fetchall()
    
    print("\n")
    for rec in results:
        print(rec)
    
    # Deleting any existing data if any from the table 'STANDINGS_13'
    if len(results):
        print("\nExisting Records Found! Deleting them now..\n")
        cur.execute("DELETE FROM STANDINGS_13")
    
    # Making a call to my_func() and saving it under 'stands'
    # Input the dates for which the standings report is required
    # In this case, 2020-12-25 for the end of season
    stands = myfunc(fixtures)
    
    # Write the data from the DataFrame 'stands' to the table 'STANDINGS_13'.
    success, nchunks, nrows, _ = write_pandas(conn, stands, 'STANDINGS_13')
    
    # Retrives and displays exported data from 'stands' from the table 'STANDINGS_13'
    results = cur.execute("SELECT * from STANDINGS_13").fetchall()
    
    print("\nExported Standings Report in Snowflake Table:")
    for rec in results:
        print(rec)

# Closing any open snowflake connections
finally:
    cur.close()
conn.close()
```
